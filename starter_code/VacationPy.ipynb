```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
gmaps.configure(api_key= g_key)
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
cities_weather_df = pd.read_csv('Cities_weather.csv')
cities_weather_df.dropna(inplace = True)
cities_weather_df.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
locations = cities_weather_df[["Lat", "Lng"]]
humidity = cities_weather_df["Humidity"].astype(float)

fig = gmaps.figure(zoom_level = 2, center = (53,46))

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights = humidity, 
                                 dissipating = False, max_intensity = 10,
                                 point_radius = 1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
cities_filtered_df = cities_weather_df[cities_weather_df['Temperature'].between(23,33)]
cities_filtered_df = cities_filtered_df[cities_filtered_df['Humidity'].between(0,30)]
cities_filtered_df = cities_filtered_df[cities_filtered_df['Cloudiness'].between(0,20)]
cities_filtered_df = cities_filtered_df[cities_filtered_df['Wind Speed'].between(0,15)]
cities_filtered_df = cities_filtered_df.drop(["Unnamed: 0"], axis = 1, inplace = False)
cities_filtered_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
cities_filtered_df['Hotel Name'] = ""
cities_filtered_df.head()
```

```python
for index, row in cities_filtered_df.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    locations = f"{lat},{lng}"
  
    params = {
    "rankby": "distance",
    "location": locations,
    "types": "lodging",
    "key": g_key
}

    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    
    try:
        response = (requests.get(url, params = params)).json()
        cities_filtered_df.loc[index, "Hotel Name"] = response["results"][0]["name"]
        print("Found hotel " + str(response["results"][0]["name"]))
       
    except (KeyError, IndexError):
        pass
```

```python
cities_filtered_df = cities_filtered_df.drop([49, 209, 258]) 
cities_filtered_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in cities_filtered_df.iterrows()]
locations = cities_filtered_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations,  hotel_info)
fig.add_layer(markers)

# Display figure
fig
```
